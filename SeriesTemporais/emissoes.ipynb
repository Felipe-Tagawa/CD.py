```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
dataset = pd.read_csv('co2_emissions.csv', delimiter=',', index_col='Year', parse_dates= True)

plt.figure()
plt.plot(dataset['CO2_Emissions'])
plt.title('CO2_Emissions per Year')
plt.xlabel('Year')
plt.ylabel('Emissions')
plt.show()

decomposition = seasonal_decompose(dataset['CO2_Emissions'], model='additive')

decomposition.plot()
plt.tight_layout()
```
