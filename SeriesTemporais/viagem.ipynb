```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
dataset = pd.read_csv('airtravel.csv', delimiter=',', index_col='Date', parse_dates= True)

plt.figure()
plt.plot(dataset['Passengers'])
plt.title('Air Travel - Passengers')
plt.xlabel('Index')
plt.ylabel('Passengers')
plt.show()

decomposition = seasonal_decompose(dataset['Passengers'], model='additive')

decomposition.plot()
plt.tight_layout()
```

R: Sim, tendncia crescente.

b. A srie possui sazonalidade? Se sim, qual o perodo que ela acontece?
